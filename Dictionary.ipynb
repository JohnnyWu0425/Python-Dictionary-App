# 1.Import Module

```python
import json      #讀取json檔案
import difflib   #找到相似字
from difflib import get_close_matches   #找到相似替換字
```

# 2.Load Data

```python
#將資料存入data中
data = json.load(open("dictionary.json"))
```

# 3.Retrieve Data

```python
#將資料依照使用者搜尋取出
def retrieve_definition(word):
    if word.title() in data.keys():
        return data[word.title()]
    elif word.upper() in data.keys():
        return data[word.upper()]
    elif word.lower() in data.keys():
        return data[word.lower()]
    
    else:
        if len(get_close_matches(word,data.keys(),n=1,cutoff=0.5))>0:
            instead_word = get_close_matches(word,data.keys(),n=1)[0]
            print("Do you mean \'%s\' instead ? [y/n]" %instead_word)
            
            yes_or_no = input()
            
            if yes_or_no == 'y':
                return(data[instead_word])
                    
            elif yes_or_no == 'n':
                return("Sorry, I don't know what you want.")
            
            else:
                return("Sorry, please enter \'y\' or \'n\'.")

        else:
            return("Sorry, I don't know what you want.")
    
    
```

# 4.Present Data

```python
#接收搜尋字
user_word = input("Please input a word you want to search: ")

print()
output = retrieve_definition(user_word)

for item in output:
    print("-"+item)
```
